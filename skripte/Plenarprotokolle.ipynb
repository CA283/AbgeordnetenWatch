```python
from selenium import webdriver
import time, re, requests
```

```python
url = 'http://pdok.bundestag.de/index.php?qsafe=%23change%23&aload=off&q=plenarprotokoll&dart=Plenarprotokoll&x=0&y=0&df=07.09.1949&dt=23.02.2017'

pages = 500
link = []

driver = webdriver.Firefox()
driver.get(url) 
time.sleep(10)

for page in range(1, pages +1):
    print ("page number: %s" % page)

    linkDiv = driver.find_elements_by_class_name("linkGeneric")
    for div in linkDiv:
        link.append(div.find_element_by_tag_name('a').get_attribute('href'))
        
    page_link = driver.find_element_by_xpath('//a[@title="eine Seite weiter"]')
    page_link.click()
    time.sleep(10)

driver.close()
```

```python
loopNumber = 1
for url in set(link):
    name = re.sub('.*/btp/', '', url)
    name = re.sub('/', '_', name)
    response = requests.get(url)
    with open('output/plenarprotokolle/PDF/' + name, 'wb') as f:
        f.write(response.content)
    print ('Loop Nr. ' + str(loopNumber) + ' done.')
    loopNumber += 1
    time.sleep(3)
```
